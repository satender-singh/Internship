```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome(r"C:\Users\yogendra\Downloads\chromedriver_win32/chromedriver.exe")
```

```python
driver.get("https://www.statisticstimes.com/")
```

```python
eco_nomy=driver.find_element(By.XPATH,"/html/body/div[2]/div[1]/div[2]/div[2]/button")
eco_nomy.click()
india=driver.find_element(By.XPATH,"/html/body/div[2]/div[1]/div[2]/div[2]/div/a[3]")
india.click()
```

```python
gdp_india=driver.find_element(By.XPATH,"/html/body/div[2]/div[2]/div[2]/ul/li[1]/a")
gdp_india.click()
```

```python
rank_id=[]
state_id=[]
gsdp_1819=[]
gsdp_1920=[]
share_1819=[]
gdp_billion=[]
```

```python
rank=driver.find_elements(By.XPATH,'//td[@class="data1"]')
for i in rank:
    rank1=i.text
    rank_id.append(rank1)
```

```python
state=driver.find_elements(By.XPATH,'//td[@class="name"]')
for i in state:
    state1=i.text
    state_id.append(state1)
```

```python
gsdp=driver.find_elements(By.XPATH,'//td[@class="data sorting_1"]')
for i in gsdp:
    gsdp1=i.text
    gsdp_1819.append(gsdp1)
```

```python
gsdp2=driver.find_elements(By.XPATH,'//td[@class="data"]')
for i in gsdp2:
    gsdp3=i.text
    gsdp_1920.append(gsdp3)
```

```python
share=driver.find_elements(By.XPATH,'//td[@class="data"]')
for i in share:
    share1=i.text
    share_1819.append(share1)
```

```python
gdp=driver.find_elements(By.XPATH,'//td[@class="data"]')
for i in gdp:
    gdp1=i.text
    gdp_billion.append(gdp1)
```

```python
print(len(rank_id),len(state_id),len(gsdp_1819),len(gsdp_1920),len(share_1819),len(gdp_billion))
```
